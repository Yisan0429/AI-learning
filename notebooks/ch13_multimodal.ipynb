# 第十三章：多模态 AI

## 从纯文本到图像、语音、视频的统一理解

**多模态 AI** 能同时处理和生成多种类型的数据：文本、图像、音频、视频。
这不再是 CV+NLP 的简单拼接，而是**统一的表示空间**中跨模态理解。

## 13.1 视觉-语言模型

### CLIP：连接图像与文本

OpenAI 的 CLIP (2021) 用**对比学习**将图像和文本映射到**共享嵌入空间**：

```
图像 → Vision Encoder → 图像向量 ┐
                                 ├→ 余弦相似度 → 匹配
文本 → Text Encoder  → 文本向量 ┘
```

CLIP 的核心创新：**4 亿图文对**的大规模弱监督训练。训练后可直接用于零样本图像分类：

```
"一张{类别}的照片" → Text Encoder → 与所有图像向量计算相似度 → 最高分即为预测类别
```

### GPT-4V / Gemini Vision

现代多模态模型不再分离视觉和文本编码器，而是将图像 token 化后直接输入 LLM：

```
图像 → ViT patch embedding → 图像 tokens + 文本 tokens → Transformer → 输出
```

这意味着同一个模型可以：看图回答问题、根据草图生成代码、分析图表数据。

## 13.2 语音处理

### 语音转文本 (STT): Whisper

OpenAI Whisper 在 68 万小时多语言数据上训练，支持 99 种语言：

| 模型 | 参数 | 相对速度 | 适用场景 |
|------|------|---------|---------|
| `tiny` | 39M | ~10x | 实时流式 |
| `base` | 74M | ~7x | 移动端 |
| `small` | 244M | ~4x | 平衡选择 |
| `medium` | 769M | ~2x | 高质量需求 |
| `large-v3` | 1550M | 1x | 最佳准确率 |

### 文本转语音 (TTS)

从传统拼接合成到端到端神经 TTS（如 VITS、XTTS），现代 TTS 可**零样本克隆**声音——只需 3-10 秒参考音频。

### 语音助手管道

```
用户语音 → Whisper STT → LLM 理解/推理 → TTS 合成 → 语音回复
```

关键挑战：**延迟**（端到端 < 500ms 才算实时对话）和**打断处理**。

## 13.3 多模态 RAG

传统 RAG 只检索文本。**多模态 RAG** 扩展了检索和生成的范围：

```
用户上传一张产品图片 + 文字问"类似款推荐"
    ↓
1. CLIP embedding 检索视觉相似的产品图片
2. 文本检索产品描述
3. 多模态 LLM (GPT-4V) 综合图片+文本 → 生成推荐
```

### 多模态 Embedding

| 模型 | 模态 | 维度 |
|------|------|------|
| `CLIP-ViT-L/14` | 图像+文本 | 768 |
| `ImageBind` | 图像/文本/音频/深度/热力/IMU | 1024 |
| `jina-clip-v1` | 图像+文本（中英） | 768 |

### 多模态 Agent

Agent 不再只能"调用搜索 API"——可以**看图、听音频、分析视频帧**：

```
Agent: 监控摄像头发现异常 → 调用视觉模型分析 → 
       判断为入侵 → 发送警报 + 截图 → 记录日志
```

## 本章小结

| 领域 | 核心技术 | 代表模型 |
|------|---------|---------|
| **视觉-语言** | 对比学习，共享嵌入空间 | CLIP, GPT-4V, Gemini Vision |
| **语音处理** | STT + TTS 端到端管道 | Whisper, XTTS |
| **多模态 RAG** | 跨模态检索 + 多模态 LLM | ImageBind, ColPali |
| **多模态 Agent** | 视觉+文本工具调用 | GPT-4V + function calling |

> 多模态 AI 正从"能做"走向"好用"——延迟、成本、可靠性是工程落地的核心瓶颈。
