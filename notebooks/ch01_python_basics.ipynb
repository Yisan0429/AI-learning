## 1.1 Python 是什么？

Python 是一种**解释型、动态类型、面向对象**的高级编程语言，由 Guido van Rossum 于 1991 年发布。

**核心特点：**
- **简洁优雅**：用缩进代替花括号，减少视觉噪音
- **动态类型**：变量无需声明类型，运行时自动推断
- **丰富的标准库**：内置模块覆盖文件 I/O、网络、正则等
- **庞大的第三方生态**：PyPI 上有超过 40 万个包

**执行模型：** Python 代码先被编译为字节码（`.pyc`），再由 Python 虚拟机（PVM）解释执行。


### Python vs C++：核心差异速览

如果你有 C++ 背景，这张表能帮你快速建立对应关系：

| 概念 | C++ | Python |
|------|-----|--------|
| **执行方式** | 编译为机器码后执行 | 解释执行（先编译为字节码 .pyc，再由虚拟机运行） |
| **类型系统** | 静态类型，编译时检查 | 动态类型，运行时检查 |
| **内存管理** | `new`/`delete`，手动或智能指针 | 引用计数 + 垃圾回收，自动管理 |
| **变量语义** | 变量是一块内存的别名 | 变量是对象的引用（名字绑定） |
| **代码块** | `{}` 花括号 | 缩进（4 空格） |
| **编译/运行** | `g++ main.cpp -o main && ./main` | `python3 main.py` 一步到位 |
| **性能** | 极致优化（裸机速度） | 慢 10-100 倍，但 AI 中计算密集型操作在 C++/CUDA 底层库（PyTorch/NumPy）中完成 |
| **典型用途** | 系统编程、游戏引擎、高频交易 | AI/ML、数据分析、Web 后端、脚本自动化 |

**一句话：Python 写起来快，C++ 跑起来快。AI 开发中 Python 做胶水，把 C++/CUDA 写的底层算子粘起来。**


````python
# 你的第一行 Python 代码
print("Hello, AI World!")

# Python 作为计算器
print(2 + 3 * 4)        # 运算优先级：先乘后加
print(2 ** 10)          # 幂运算：2^10 = 1024
print(17 // 5, 17 % 5)  # 整除 3，取余 2
````

## 1.2 变量与基本数据类型

在 Python 中，变量是**对象的引用**（名字绑定），而非存储数据的容器。赋值 `=` 将名字绑定到对象上。

### 基本数据类型

| 类型 | 示例 | 说明 |
|------|------|------|
| `int` | `42`, `-7`, `0b1010` | 整数（任意精度） |
| `float` | `3.14`, `-2.5e-3` | 双精度浮点数（IEEE 754） |
| `str` | `"hello"`, `'世界'` | 不可变 Unicode 字符串 |


````python
# === 变量赋值与类型检查 ===
x = 42                          # int
pi = 3.14159                    # float
name = "人工智能"                # str
is_valid = True                 # bool
nothing = None                  # NoneType

print(type(x), type(pi), type(name), type(is_valid), type(nothing))

# 动态类型：变量可以随时重新绑定到不同类型的对象
x = "现在是字符串了"
print(type(x))

# 多重赋值
a, b, c = 1, 2.0, "三"
print(a, b, c)

# 交换变量（Python 特有语法，无需临时变量）
a, b = b, a
print(f"交换后: a={a}, b={b}")
````

````python
# === for 循环 ===
# range(start, stop, step)：生成整数序列（左闭右开）
print("range(5):", list(range(5)))        # [0,1,2,3,4]
print("range(2,8,2):", list(range(2,8,2))) # [2,4,6]

# 遍历列表（带索引：enumerate）
fruits = ["苹果", "香蕉", "橘子"]
for i, fruit in enumerate(fruits):
    print(f"  {i}: {fruit}")

# 同时遍历两个列表（zip）
names = ["Alice", "Bob", "Charlie"]
scores = [95, 87, 92]
for name, score in zip(names, scores):
    print(f"  {name}: {score}分")
````

````python
# === while 循环 ===
# 经典：梯度下降模拟
x = 10.0          # 初始值
lr = 0.1          # 学习率
step = 0

while x > 0.01:   # 当 x 足够小時停止
    x = x - lr * 2 * x  # f(x) = x^2 的梯度下降
    step += 1
    if step <= 5:
        print(f"  Step {step}: x = {x:.4f}")

print(f"共 {step} 步，最终 x = {x:.6f}")

# break 和 continue
for i in range(10):
    if i == 3:
        continue  # 跳过本次
    if i == 7:
        break     # 终止循环
    print(i, end=" ")
print()  # 输出: 0 1 2 4 5 6
````

## 1.5 函数

函数是组织代码的基本单元。Python 使用 `def` 关键字定义函数。

### 函数定义语法
```python
def function_name(param1, param2=default):
    """文档字符串 (docstring)"""
    # 函数体
    return result
```


````python
# === 基础函数 ===
import math

def sigmoid(x):
    """Sigmoid 激活函数: f(x) = 1 / (1 + e^(-x))"""
    return 1 / (1 + math.exp(-x))  # 返回结果

print(f"sigma(0)  = {sigmoid(0):.4f}")   # 0.5000
print(f"sigma(2)  = {sigmoid(2):.4f}")   # 0.8808
print(f"sigma(-2) = {sigmoid(-2):.4f}")  # 0.1192

# === 默认参数 ===
def create_optimizer(name="Adam", lr=0.001, weight_decay=0.0):
    return {"name": name, "lr": lr, "weight_decay": weight_decay}  # 返回结果

print(create_optimizer())                        # 全部默认
print(create_optimizer(lr=0.01, name="SGD"))     # 关键字参数（顺序无关）
````

````python
# === 可变参数：*args 和 **kwargs ===
def summarize(*args, **kwargs):
    """
    *args:  接收任意数量的位置参数（打包为元组）
    **kwargs: 接收任意数量的关键字参数（打包为字典）
    """
    print(f"位置参数 {len(args)} 个: {args}")
    print(f"关键字参数 {len(kwargs)} 个: {kwargs}")
    return sum(args) if args else 0  # 返回结果

total = summarize(1, 2, 3, name="训练", epoch=10)
print(f"总和: {total}")

# lambda 表达式：匿名函数，适合简单操作
square = lambda x: x ** 2
add = lambda a, b: a + b
print(f"lambda: square(5)={square(5)}, add(2,3)={add(2,3)}")

# 常见用法：作为 sorted/map/filter 的 key
pairs = [(1, "one"), (3, "three"), (2, "two")]
sorted_pairs = sorted(pairs, key=lambda x: x[0])
print(f"排序后: {sorted_pairs}")
````

## 1.6 面向对象编程 (OOP)

Python 万物皆对象。通过 `class` 定义自定义类型，支持封装、继承、多态。

### 核心概念

| 概念 | Python 实现 |
|------|-------------|
| 构造函数 | `__init__(self)` |
| 字符串表示 | `__str__` / `__repr__` |
| 实例方法 | 第一个参数必须是 `self` |
| 类变量 | 在 `__init__` 外定义，所有实例共享 |
| 继承 | `class Child(Parent)` |
| 多态 | 鸭子类型：如果它走路像鸭子，叫起来像鴨子，那它就是鸭子 |


````python
# === 基础类：定义一个简单的神经网络层 ===
import math, random

class DenseLayer:
    """全连接层：y = activation(W * x + b)"""
    
    def __init__(self, input_dim, output_dim, activation="relu"):
        self.W = [[random.gauss(0, 0.1) for _ in range(input_dim)] 
                  for _ in range(output_dim)]
        self.b = [0.0] * output_dim
        self.activation = activation
        self.input_dim = input_dim
        self.output_dim = output_dim
    
    def __repr__(self):
        return f"DenseLayer({self.input_dim} -> {self.output_dim}, {self.activation})"  # 返回结果
    
    def forward(self, x):
        """前向传播（简化版）"""
        y = [sum(self.W[j][i] * x[i] for i in range(self.input_dim)) + self.b[j]
             for j in range(self.output_dim)]
        return y  # 返回结果
    
    @property
    def num_params(self):
        """参数数量"""
        return self.input_dim * self.output_dim + self.output_dim  # 返回结果

# 创建和使用
layer1 = DenseLayer(784, 128, activation="relu")
layer2 = DenseLayer(128, 10, activation="softmax")

print(layer1)
print(layer2)
print(f"Layer 1 参数数: {layer1.num_params:,}")
print(f"Layer 2 参数数: {layer2.num_params:,}")

# 模拟一个输入
x = [0.5] * 784
h = layer1.forward(x)
print(f"隐藏层输出前5维: {[round(v, 4) for v in h[:5]]}")
````

````python
# === 继承与多态 ===
class Optimizer:
    """优化器基类"""
    def __init__(self, lr=0.01):
        self.lr = lr
    
    def step(self, params, grads):
        raise NotImplementedError("子类必须实现 step 方法")

class SGD(Optimizer):
    """随机梯度下降"""
    def step(self, params, grads):
        return [p - self.lr * g for p, g in zip(params, grads)]  # 返回结果

class Momentum(Optimizer):
    """带动量的 SGD"""
    def __init__(self, lr=0.01, momentum=0.9):
        super().__init__(lr)
        self.momentum = momentum
        self.velocity = None
    
    def step(self, params, grads):
        if self.velocity is None:
            self.velocity = [0.0] * len(params)
        self.velocity = [self.momentum * v - self.lr * g 
                         for v, g in zip(self.velocity, grads)]
        return [p + v for p, v in zip(params, self.velocity)]  # 返回结果

# 多态：统一接口，不同行为
def train_step(optimizer, params, grads):
    return optimizer.step(params, grads)  # 返回结果

params = [1.0, 2.0, 3.0]
grads = [0.1, 0.2, 0.3]

sgd = SGD(lr=0.1)
momentum = Momentum(lr=0.1, momentum=0.9)

print(f"SGD 一步: {train_step(sgd, params, grads)}")
print(f"Momentum 一步: {train_step(momentum, params, grads)}")
````

````python
# === 异常处理 ===
def safe_divide(a, b):
    """try/except/else/finally 完整模式"""
    try:
        result = a / b
    except ZeroDivisionError:
        print(f"错误: 除数为零 ({a}/{b})")
        return float('inf')  # 返回结果
    except TypeError as e:
        print(f"类型错误: {e}")
        return None  # 返回结果
    else:
        print(f"计算成功: {a}/{b} = {result:.4f}")
        return result  # 返回结果
    finally:
        pass  # 无论是否异常都执行（常用于资源清理）

print(safe_divide(10, 3))
print(safe_divide(10, 0))
print(safe_divide("10", 3))
````

## 1.8 综合练习：实现 Mini DataLoader

综合运用本章知识点，实现一个简化版的数据批处理迭代器，为后续深度学习章节做准备。


````python
# === 综合练习：Batch DataLoader ===
import random

class DataLoader:
    """简化版数据加载器：实现 shuffle + batch + 迭代"""
    
    def __init__(self, data, labels, batch_size=32, shuffle=True):
        self.data = data
        self.labels = labels
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.n_samples = len(data)
    
    def __len__(self):
        """返回 batch 数量"""
        return (self.n_samples + self.batch_size - 1) // self.batch_size  # 返回结果
    
    def __iter__(self):
        """迭代器协议：生成 (batch_data, batch_labels)"""
        indices = list(range(self.n_samples))
        if self.shuffle:
            random.shuffle(indices)
        
        for start in range(0, self.n_samples, self.batch_size):
            batch_idx = indices[start:start + self.batch_size]
            yield [self.data[i] for i in batch_idx], [self.labels[i] for i in batch_idx]

# 模拟数据：100 条样本，每条 4 个特征
mock_data = [[random.random() for _ in range(4)] for _ in range(100)]
mock_labels = [random.randint(0, 2) for _ in range(100)]  # 3 分类

loader = DataLoader(mock_data, mock_labels, batch_size=16)
print(f"总样本: {loader.n_samples}, Batch 数: {len(loader)}, 每 Batch: {loader.batch_size}")

# 遍历 2 个 batch
for i, (batch_x, batch_y) in enumerate(loader):
    print(f"Batch {i}: {len(batch_x)} 条样本, 标签分布: "
          f"{{0: {batch_y.count(0)}, 1: {batch_y.count(1)}, 2: {batch_y.count(2)}}}")
    if i >= 1:
        break

print("\n✅ 第一章完成！你已掌握 Python 核心语法。")
````
