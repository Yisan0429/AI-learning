## 1.1 Python 是什么？

Python 是一种**解释型、动态类型、面向对象**的高级编程语言，由 Guido van Rossum 于 1991 年发布。

**核心特点：**
- **简洁优雅**：用缩进代替花括号，减少视觉噪音
- **动态类型**：变量无需声明类型，运行时自动推断
- **丰富的标准库**：内置模块覆盖文件 I/O、网络、正则等
- **庞大的第三方生态**：PyPI 上有超过 40 万个包

**执行模型：** Python 代码先被编译为字节码（`.pyc`），再由 Python 虚拟机（PVM）解释执行。


### Python vs C++：核心差异速览

如果你有 C++ 背景，这张表能帮你快速建立对应关系：

| 概念 | C++ | Python |
|------|-----|--------|
| **执行方式** | 编译为机器码后执行 | 解释执行（先编译为字节码 .pyc，再由虚拟机运行） |
| **类型系统** | 静态类型，编译时检查 | 动态类型，运行时检查 |
| **内存管理** | `new`/`delete`，手动或智能指针 | 引用计数 + 垃圾回收，自动管理 |
| **变量语义** | 变量是一块内存的别名 | 变量是对象的引用（名字绑定） |
| **代码块** | `{}` 花括号 | 缩进（4 空格） |
| **编译/运行** | `g++ main.cpp -o main && ./main` | `python3 main.py` 一步到位 |
| **性能** | 极致优化（裸机速度） | 慢 10-100 倍，但 AI 中计算密集型操作在 C++/CUDA 底层库（PyTorch/NumPy）中完成 |
| **典型用途** | 系统编程、游戏引擎、高频交易 | AI/ML、数据分析、Web 后端、脚本自动化 |

**一句话：Python 写起来快，C++ 跑起来快。AI 开发中 Python 做胶水，把 C++/CUDA 写的底层算子粘起来。**


````python
# 你的第一行 Python 代码
print("Hello, AI World!")

# Python 作为计算器
print(2 + 3 * 4)        # 运算优先级：先乘后加
print(2 ** 10)          # 幂运算：2^10 = 1024
print(17 // 5, 17 % 5)  # 整除 3，取余 2
````

## 1.2 变量与基本数据类型

在 Python 中，变量是**对象的引用**（名字绑定），而非存储数据的容器。赋值 `=` 将名字绑定到对象上。

### 基本数据类型

| 类型 | 示例 | 说明 |
|------|------|------|
| `int` | `42`, `-7`, `0b1010` | 整数（任意精度） |

````python
# === while 循环 ===
# 经典：梯度下降模拟
x = 10.0          # 初始值
lr = 0.1          # 学习率
step = 0

while x > 0.01:   # 当 x 足够小時停止
    x = x - lr * 2 * x  # f(x) = x^2 的梯度下降
    step += 1
    if step <= 5:
        print(f"  Step {step}: x = {x:.4f}")

print(f"共 {step} 步，最终 x = {x:.6f}")

# break 和 continue
for i in range(10):
    if i == 3:
        continue  # 跳过本次
    if i == 7:
        break     # 终止循环
    print(i, end=" ")
print()  # 输出: 0 1 2 4 5 6
````

## 1.5 函数

函数是组织代码的基本单元。Python 使用 `def` 关键字定义函数。

### 函数定义语法
```python
def function_name(param1, param2=default):
    """文档字符串 (docstring)"""
    # 函数体
    return result
```


````python
# === 基础函数 ===
import math

def sigmoid(x):
    """Sigmoid 激活函数: f(x) = 1 / (1 + e^(-x))"""
    return 1 / (1 + math.exp(-x))  # 返回结果

print(f"sigma(0)  = {sigmoid(0):.4f}")   # 0.5000
print(f"sigma(2)  = {sigmoid(2):.4f}")   # 0.8808
print(f"sigma(-2) = {sigmoid(-2):.4f}")  # 0.1192

# === 默认参数 ===
def create_optimizer(name="Adam", lr=0.001, weight_decay=0.0):
    return {"name": name, "lr": lr, "weight_decay": weight_decay}  # 返回结果

print(create_optimizer())                        # 全部默认
print(create_optimizer(lr=0.01, name="SGD"))     # 关键字参数（顺序无关）
````

````python
# === 可变参数：*args 和 **kwargs ===
def summarize(*args, **kwargs):
    """
    *args:  接收任意数量的位置参数（打包为元组）
    **kwargs: 接收任意数量的关键字参数（打包为字典）
    """
    print(f"位置参数 {len(args)} 个: {args}")
    print(f"关键字参数 {len(kwargs)} 个: {kwargs}")
    return sum(args) if args else 0  # 返回结果

total = summarize(1, 2, 3, name="训练", epoch=10)
print(f"总和: {total}")

# lambda 表达式：匿名函数，适合简单操作
square = lambda x: x ** 2
add = lambda a, b: a + b
print(f"lambda: square(5)={square(5)}, add(2,3)={add(2,3)}")

# 常见用法：作为 sorted/map/filter 的 key
pairs = [(1, "one"), (3, "three"), (2, "two")]
sorted_pairs = sorted(pairs, key=lambda x: x[0])
print(f"排序后: {sorted_pairs}")
````

## 1.6 面向对象编程 (OOP)

Python 万物皆对象。通过 `class` 定义自定义类型，支持封装、继承、多态。

### 核心概念

| 概念 | Python 实现 |
|------|-------------|
| 构造函数 | `__init__(self)` |
| 字符串表示 | `__str__` / `__repr__` |
| 实例方法 | 第一个参数必须是 `self` |
| 类变量 | 在 `__init__` 外定义，所有实例共享 |
| 继承 | `class Child(Parent)` |
| 多态 | 鸭子类型：如果它走路像鸭子，叫起来像鴨子，那它就是鸭子 |


````python
# === 基础类：定义一个简单的神经网络层 ===
import math, random

class DenseLayer:
    """全连接层：y = activation(W * x + b)"""
    
    def __init__(self, input_dim, output_dim, activation="relu"):
        self.W = [[random.gauss(0, 0.1) for _ in range(input_dim)] 
                  for _ in range(output_dim)]
        self.b = [0.0] * output_dim
        self.activation = activation
        self.input_dim = input_dim
        self.output_dim = output_dim
    
    def __repr__(self):
        return f"DenseLayer({self.input_dim} -> {self.output_dim}, {self.activation})"  # 返回结果
    
    def forward(self, x):
        """前向传播（简化版）"""
        y = [sum(self.W[j][i] * x[i] for i in range(self.input_dim)) + self.b[j]
             for j in range(self.output_dim)]
        return y  # 返回结果
    
    @property
    def num_params(self):
        """参数数量"""
        return self.input_dim * self.output_dim + self.output_dim  # 返回结果

# 创建和使用
layer1 = DenseLayer(784, 128, activation="relu")
layer2 = DenseLayer(128, 10, activation="softmax")

print(layer1)
print(layer2)
print(f"Layer 1 参数数: {layer1.num_params:,}")
print(f"Layer 2 参数数: {layer2.num_params:,}")

# 模拟一个输入
x = [0.5] * 784
h = layer1.forward(x)
print(f"隐藏层输出前5维: {[round(v, 4) for v in h[:5]]}")
````

````python
# === 继承与多态 ===
class Optimizer:
    """优化器基类"""
    def __init__(self, lr=0.01):
        self.lr = lr
    
    def step(self, params, grads):
        raise NotImplementedError("子类必须实现 step 方法")

class SGD(Optimizer):
    """随机梯度下降"""
    def step(self, params, grads):
        return [p - self.lr * g for p, g in zip(params, grads)]  # 返回结果

class Momentum(Optimizer):
    """带动量的 SGD"""
    def __init__(self, lr=0.01, momentum=0.9):
        super().__init__(lr)
        self.momentum = momentum
        self.velocity = None
    
    def step(self, params, grads):
        if self.velocity is None:
            self.velocity = [0.0] * len(params)
        self.velocity = [self.momentum * v - self.lr * g 
                         for v, g in zip(self.velocity, grads)]
        return [p + v for p, v in zip(params, self.velocity)]  # 返回结果

# 多态：统一接口，不同行为
def train_step(optimizer, params, grads):
    return optimizer.step(params, grads)  # 返回结果

params = [1.0, 2.0, 3.0]
grads = [0.1, 0.2, 0.3]

sgd = SGD(lr=0.1)
momentum = Momentum(lr=0.1, momentum=0.9)

print(f"SGD 一步: {train_step(sgd, params, grads)}")
print(f"Momentum 一步: {train_step(momentum, params, grads)}")
````

````python
# === 异常处理 ===
def safe_divide(a, b):
    """try/except/else/finally 完整模式"""
    try:
        result = a / b
    except ZeroDivisionError:
        print(f"错误: 除数为零 ({a}/{b})")
        return float('inf')  # 返回结果
    except TypeError as e:
        print(f"类型错误: {e}")
        return None  # 返回结果
    else:
        print(f"计算成功: {a}/{b} = {result:.4f}")
        return result  # 返回结果
    finally:
        pass  # 无论是否异常都执行（常用于资源清理）

print(safe_divide(10, 3))
print(safe_divide(10, 0))
print(safe_divide("10", 3))
````

## 1.8 综合练习：实现 Mini DataLoader

综合运用本章知识点，实现一个简化版的数据批处理迭代器，为后续深度学习章节做准备。


````python
# === 综合练习：Batch DataLoader ===
import random

class DataLoader:
    """简化版数据加载器：实现 shuffle + batch + 迭代"""
    
    def __init__(self, data, labels, batch_size=32, shuffle=True):
        self.data = data
        self.labels = labels
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.n_samples = len(data)
    
    def __len__(self):
        """返回 batch 数量"""
        return (self.n_samples + self.batch_size - 1) // self.batch_size  # 返回结果
    
    def __iter__(self):
        """迭代器协议：生成 (batch_data, batch_labels)"""
        indices = list(range(self.n_samples))
        if self.shuffle:
            random.shuffle(indices)
        
        for start in range(0, self.n_samples, self.batch_size):
            batch_idx = indices[start:start + self.batch_size]
            yield [self.data[i] for i in batch_idx], [self.labels[i] for i in batch_idx]

# 模拟数据：100 条样本，每条 4 个特征
mock_data = [[random.random() for _ in range(4)] for _ in range(100)]
mock_labels = [random.randint(0, 2) for _ in range(100)]  # 3 分类

loader = DataLoader(mock_data, mock_labels, batch_size=16)
print(f"总样本: {loader.n_samples}, Batch 数: {len(loader)}, 每 Batch: {loader.batch_size}")

# 遍历 2 个 batch
for i, (batch_x, batch_y) in enumerate(loader):
    print(f"Batch {i}: {len(batch_x)} 条样本, 标签分布: "
          f"{{0: {batch_y.count(0)}, 1: {batch_y.count(1)}, 2: {batch_y.count(2)}}}")
    if i >= 1:
        break

print("\n✅ 第一章完成！你已掌握 Python 核心语法。")
````

## 1.9 Python 环境与包管理

## 环境隔离、依赖管理、可复现的 AI 开发工作流

AI 项目依赖大量第三方库（PyTorch, Transformers, NumPy 等），不同项目可能需要**不同版本**的同一库。

### 1.9.1 为什么需要虚拟环境？

### 问题场景

假设你同时维护两个项目：

| 项目 | 需要 PyTorch | 需要 NumPy |
|------|-------------|------------|
| 项目 A（旧） | 1.13.0 | 1.23.0 |
| 项目 B（新） | 2.5.0 | 2.0.0 |

如果全局安装，PyTorch 和 NumPy 只能有一个版本——必然有一个项目无法运行。

### 核心概念

```
全局 Python (系统级)
├── venv_project_a/     ← 独立的 Python + site-packages
│   ├── torch==1.13.0
│   └── numpy==1.23.0
├── venv_project_b/     ← 完全隔离，互不影响
│   ├── torch==2.5.0
│   └── numpy==2.0.0
└── conda_env_dl/       ← Conda 环境，可装非 Python 依赖 (CUDA, C++ libs)
    ├── pytorch=2.5.0
    └── cudatoolkit=12.1
```

### 三类主流方案

| 工具 | 适用场景 | 特点 |
|------|---------|------|
| `venv` | 纯 Python 项目 | Python 内置，轻量，仅管理 Python 包 |
| `virtualenv` | 需要 Python 版本切换 | 比 venv 功能更丰富 |
| `conda` | AI/科学计算 | 管理 Python + 非 Python 依赖（CUDA, MKL, C++ libs）|


### 1.9.2 venv — Python 内置虚拟环境

`venv` 从 Python 3.3 开始内置，无需安装。创建虚拟环境会复制一个**轻量级 Python 解释器**和独立的 `site-packages` 目录。


````python
# === 在终端中执行（非 notebook cell）===
# 
# # 1. 创建虚拟环境
# python3 -m venv myenv
#
# # 2. 激活环境
# # Linux/macOS:
# source myenv/bin/activate
# # Windows:
# myenv\Scripts\activate
#
# # 3. 确认环境
# which python    # 应指向 myenv/bin/python
# pip list        # 应为空（仅 pip 和 setuptools）
#
# # 4. 安装依赖
# pip install numpy pandas matplotlib
#
# # 5. 导出依赖列表
# pip freeze > requirements.txt
#
# # 6. 退出环境
# deactivate
````

````python
# === 在 Python 中检查当前环境 ===
import sys
import os

print(f"Python 路径: {sys.executable}")
print(f"是否虚拟环境: {sys.prefix != sys.base_prefix}")
print(f"site-packages: {sys.path[-3:]}")

# 查看已安装包的版本
import importlib.metadata as im

packages_to_check = ["numpy", "pandas", "matplotlib"]
for pkg in packages_to_check:
    try:
        version = im.version(pkg)
        print(f"  {pkg}=={version}")
    except im.PackageNotFoundError:
        print(f"  {pkg}: 未安装")
````

### 1.9.4 Conda — 数据科学/AI 的首选环境管理器

### Conda vs pip

| | pip | conda |
|---|-----|-------|
| 安装内容 | 仅 Python 包 | Python 包 + C/C++/R 库 |
| 依赖解析 | 较简单（可能冲突） | SAT solver，更严格 |
| 环境管理 | 需配合 venv | 内置环境管理 |
| 包来源 | PyPI | Anaconda Cloud / conda-forge |
| CUDA 管理 | 手动安装 | `cudatoolkit` 一键安装 |
| 典型场景 | Web 开发、脚本 | 数据科学、深度学习、生物信息 |


````python
# === Conda 常用命令（在终端执行）===
#
# # ---- 环境管理 ----
# conda create -n myenv python=3.11        # 创建环境
# conda activate myenv                     # 激活环境
# conda deactivate                         # 退出环境
# conda env list                           # 列出所有环境
# conda remove -n myenv --all              # 删除环境
# conda env export > environment.yml       # 导出环境
# conda env create -f environment.yml      # 从文件创建环境
#
# # ---- 包管理 ----
# conda install numpy pandas               # 安装包
# conda install pytorch torchvision \
#   pytorch-cuda=12.1 -c pytorch -c nvidia # 带 channel 安装
# conda list                               # 列出已安装包
# conda update --all                       # 更新所有包
# conda remove numpy                       # 卸载包
#
# # ---- 配置 ----
# conda config --show channels             # 查看 channel
# conda config --add channels conda-forge  # 添加 channel
````

### Conda channel 与优先级

Conda 从 **channel**（软件仓库）下载包。默认 channel 是 Anaconda 官方仓库。

**推荐 channel 配置顺序：**

`conda-forge` 是社区维护的最大 channel，包更新更快、更全。

### 1.9.5 requirements.txt — 精确锁定依赖版本

`requirements.txt` 是 Python 项目的**依赖清单**，确保任何人在任意机器上可以通过 `pip install -r requirements.txt` 复现完全相同的环境。


````python
# === 生成 requirements.txt ===
# 
# 方式 1：导出当前环境所有包（最常用）
# pip freeze > requirements.txt
#
# 方式 2：使用 pip-tools 精确管理（推荐正式项目）
# pip install pip-tools
# # 编辑 requirements.in（只写直接依赖）
# # numpy
# # torch>=2.0
# # transformers
# pip-compile requirements.in  # 生成锁定的 requirements.txt

# === 示例：一个典型 AI 项目的 requirements.txt ===
sample_requirements = """
# Core
numpy>=1.24,<2.0
scipy>=1.10
pandas>=2.0

# Deep Learning
torch>=2.0
torchvision>=0.15

# NLP
transformers>=4.30
tokenizers>=0.13
datasets>=2.12

# Visualization
matplotlib>=3.7
seaborn>=0.12

# Dev Tools
jupyter>=1.0
ipykernel>=6.0
"""

with open("/tmp/sample_requirements.txt", "w") as f:
    f.write(sample_requirements)

print("示例 requirements.txt:")
print(sample_requirements)
import os; os.remove("/tmp/sample_requirements.txt")
````

### 版本号规范

```python
torch==2.5.0       # 精确版本（最安全但不够灵活）
torch>=2.0         # 至少 2.0，允许更高
torch>=2.0,<3.0    # 兼容范围（推荐——明确上下界）
torch~=2.5.0       # 兼容版本（>=2.5.0, <2.6.0）
```

### 1.9.6 environment.yml — Conda 的依赖描述文件

`environment.yml` 比 `requirements.txt` 更强大：可以指定 Python 版本、channel，还可以包含非 Python 依赖（如 CUDA）。

````python
# === 依赖检查脚本 ===
"""
在项目根目录运行此脚本，检查关键依赖是否正确安装。
"""

import sys

def check_package(name, min_version=None):
    try:
        import importlib.metadata as im
        version = im.version(name)
        if min_version:
            from packaging import version as pv
            if pv.parse(version) < pv.parse(min_version):
                return f"⚠ {name}=={version} (require >={min_version})"  # 返回结果
        return f"✅ {name}=={version}"  # 返回结果
    except ImportError:
        return f"❌ {name}: 未安装"  # 返回结果

# 检查列表
deps = [
    ("numpy", "1.24"),
    ("pandas", "2.0"),
    ("matplotlib", "3.7"),
    ("torch", "2.0"),
    ("torchvision", "0.15"),
    ("transformers", "4.30"),
]

print("Python:", sys.version.split()[0])
print("-" * 50)
for name, min_ver in deps:
    print(check_package(name, min_ver))
````

```bash
pipx install jupyter
```

```python
import torch
print(torch.cuda.is_available())  # True 表示 GPU 可用
```
